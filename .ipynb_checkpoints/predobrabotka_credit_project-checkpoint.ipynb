<h3 style="color:green">Корректировка количества дней стажа и заполнение пропусков
</h3>

```python
# положительные значения больше 30 тысяч будут переведены в дни делением на 24
def fix_days(row):
    days = row['days_employed']
    if days > 30000:
        days_fix = days / 24
        return days_fix
    if days < 0:
        days_fix = abs(days)
        return days_fix


df['days_employed'] = df.apply(fix_days, axis=1)
```

Стаж напрямую коррелируется только с возрастом, значит эту часть данных с аномальным стажем можно скорректировать только по возрасту заёмщика. Для этого нужно вычесть из возраста заёмщика период, в который большинство людей не начинает официальную трудовую деятельность. В Российской Федерации этот период 18 лет.

<h3 style="color:green">Замена нулевых значений возраста.
</h3>

Для того, чтобы заполнить пропуски со стажем, нужно разобраться с возрастом заёмщиков

```python
# построим визуализацию распределения возраста заёмщиков
df['dob_years'].hist()
print('Количество нулевых значений возраста :',len(df[df['dob_years'] == 0]))
```

```python
#  изучим какие данные у нас есть для нулевых значений стажа
df[df['dob_years'] == 0].describe()
```

так как для нулевых значений возраста есть значения стажа, то можно заменить нулевые значения на вычисленный благодаря стажу возраст. Для этого возьмём все строки с нулевым возрастом и ненулевыми значениями стажа

```python

def fix_age(row):
    age = row['dob_years']
    days = row['days_employed']
    if age == 0 and days > 0:
        age_return = int((days / 365) + 18)
    else:
        age_return = age
        
    return age_return

df['dob_years'] = df.apply(fix_age, axis = 1)
```

```python
# осталось несколько строк с нулевым возрастом, заменим их на медианное значение возраста
print(int(df['dob_years'].median()))
df['dob_years'] = df['dob_years'].replace(0, 43)
```

<h3 style="color:green">Стаж в годах и заполнение пропусков стажа.
</h3>

Теперь для обработки аномальных значений стажа всё есть, для упрощения работы дополнительно создадим столбец в годах

```python
# порог для корректировки значений стажа будет проходить из расчёта возраст - 18 лет, так как стаж в днях, 
# значение нужно умножить на количество дней в году
def fix_days_too_big(row):
    days = row['days_employed']
    age = row['dob_years']
    if days > (age - 18) * 365:
        days_fix = (age - 18) * 365
        return days_fix
    else:
        days_fix = days
        return days_fix


df['days_employed'] = df.apply(fix_days_too_big, axis=1)
```

```python
# заполним пропуски медианным значеним стажа работы для конкретного возраста. 
def fill_days(row):
    days = row['days_employed']
    age = row['dob_years']
    if days > 0:
        days_employed = days
        return days_employed
    else:
        days_employed = df[df['dob_years'] == age]['days_employed'].median()
        return days_employed
    
df['days_employed'] = df.apply(fill_days, axis=1)
df['days_employed'] = df['days_employed'].astype(int)
```

```python
# для удобства создадим столбец со стажем в годах
df['years_employed'] = df['days_employed'] / 365
```

```python
# изучим как теперь данные по возрасту и стажу распределились на графике
sns.scatterplot(x="dob_years", y="years_employed",hue = 'gender', data=df)
```

<h3 style="color:green">Лемматизация
</h3>

Для того, чтобы сократить количество категорий для цели кредита, воспользуемся лемматизацией. Так схожие по смыслу цели можно будет объединить.

В нашем примере вариантов целей кредита не много, можно было заменить такие строки и вручную, но для ситуаций, когда вариантов одной темы, слова и т.д. много гораздо быстрей и удобней использовать лемматизацию

```python
# ещё раз вернёмя к вариантам целей
df['purpose'].value_counts()
```

Видно, что можно выделить несколько характерных целей: Недвижимость, автомобиль, образование и свадьба. Для обращения к этим названиям, нужно привести слова к начальной форме

```python
# каждую запись цели кредита представляем как список слов и к каждому элементу этого списка 
# обращаемся с запросом получить начальную форму слова
def lemma_purpose(row):
#     split_purpose это список слов одной записи о цели кредита
    split_purpose = row['purpose'].split()
    words = []
#     перебираем слова в наших списках
    for word in split_purpose:     
#         для каждого слова находим начальную форму и добавляем в список
        words.append(morph.parse(word)[0].normal_form)
#         превращаем список в строковую запись из слов начальной формы
    return ' '.join(words)
# применяем функцию к столбцу с целями кредита
df['lemms'] = df.apply(lemma_purpose, axis=1)
```

```python
# ещё раз смотрим какие ключевые слова для целей можно выделить
# df['lemms'].value_counts()
```

```python
# Теперь можно выполнить проверку для каждой строки и в случае если начальная форма слова совпадает с нашим запросом
# возвращать сокращенную версию цели
def purpose(row):
    lemma_purposes = row['lemms']
    if 'недвижимость' in lemma_purposes:
        return 'недвижимость'
    if 'жильё' in lemma_purposes:
        return 'недвижимость'
    if 'автомобиль' in lemma_purposes:
        return 'автомобиль'
    if 'образование' in lemma_purposes:
        return 'образование'
    if 'операция' in lemma_purposes:
        return 'недвижимость'
    if 'свадьба' in lemma_purposes:
        return 'свадьба'
    if 'строительство' in lemma_purposes:
        return 'строительство'
    if 'ремонт' in lemma_purposes:
        return 'строительство'
    
    
    return 'что_то_другое'
 
df['purpose_classified'] = df.apply(purpose, axis=1)
print(df['purpose_classified'].value_counts())
print(df['purpose_classified'].value_counts().sum())
# выведем три строки нового столбца с целями
df['purpose_classified'].head(3)
```

Для недвижимости были две леммы: 'жилье' и недвижимость',  'ремонт' и 'строительство' также выделены в эту цель. Так сделано по причине общего объекта собственности - недвижимости. Данную логику для объединения лучше дополнительно согласовать с представителями кредитного отдела.

Категории целей объединены по смыслу, теперь будет легче анализировать данные с ними связанные.

<h3 style="color:green">Категоризация уровня дохода заёмщика
</h3>

Доход заёмщика - важный параметр, который будет учитывать скорринговая система, для анализа данных объединим заёмщиков по уровню дохода, для этого оценим распределение этой величины.

```python
fig = plt.figure(figsize=(15,5))
sns.boxplot(df['total_income'])
plt.xlim(0,400000)
```

Заёмщиков можно разделить на тех чей доход ниже среднего, средний, выше среднего и высокий

Ниже среднего - меньше 100 000 единиц
Средний - 100 000 - 200 000 единиц
Выше среднего - 200 000 - 350 000 единиц
Высокий - больше 350 000 единиц

```python
q_1 = df['total_income'].quantile(0.25)
q_2 = df['total_income'].quantile(0.75)
print('25-ый процентиль составляет',q_1)
print('75-ый процентиль составляет',q_2)
```

```python
# 25 и 75 процентили хорошо описывают три категории
# 100 000 и меньше, заёмщики, доход которых ниже самых распространенных значение
# 100 000 - 200 000  - самый распространённый доход
# 200 000 + это доход, который выше распространенных значений
# создадим функцию с границами категорий
def group_income(row):
    total_income = row['total_income']
    if total_income < q_1:
        return 25        
    if total_income >= q_1 and total_income <= q_2:
        return 50
    if total_income > q_2:
        return 75

df['group_income'] = df.apply(group_income, axis=1)
print(df['group_income'].value_counts()) 
#df.drop_duplicates().reset_index(drop = True)
#print(df.info())
```

<h3 style="color:green">Категоризация количества детей заёмщика
</h3>


Доля заёмщиков, у которых двое детей и больше меньше других, поэтому их можно объединить в одну группу

```python
def group_сhildren(row):
    children = row['children']
    if children == 0:
        return '0'        
    if children == 1:
        return '1'
    if children >= 2:
        return '2+'


df['children_group'] = df.apply(group_сhildren, axis=1)
print(df['children_group'].value_counts()) 
```

```python
print(df['dob_years'].quantile(0.25))
print(df['dob_years'].quantile(0.75))

age_33 = df['dob_years'].quantile(0.25)
age_53 = df['dob_years'].quantile(0.75)
```

```python
def dob_years_group(row):
    age = row['dob_years']
    if age < age_33:
        return '33'        
    if age >= age_33 and age <= age_53:
        return '53'
    if age > age_53:
        return '53+'

df['years_group'] = df.apply(dob_years_group, axis=1)
print(df['years_group'].value_counts()) 
```

```python
df_debt_children = df.pivot_table(index = 'children_group', columns = 'debt', aggfunc = {'debt' : 'count'})
df_debt_children['total_credits'] = df['children_group'].value_counts()
df_debt_children['debt_percent'] = (df_debt_children['debt',  1] /  df_debt_children['total_credits']) * 100
df_debt_children.set_axis(['нет задолженностей','есть задолженность','кол-во кредитов','процент должников'],
                          axis = 'columns',inplace = True)

plt.figure(figsize=(13, 9))
ax = sns.heatmap(df.pivot_table(index = ['children_group'], columns = 'gender',values = 'debt'),
                 annot = True, fmt  = '.2%', vmin = 0.05)

plt.yticks(fontsize = 15, rotation = 0)
plt.xticks(fontsize = 15, rotation = 0)
plt.xlabel('Пол', fontsize = 20)
plt.ylabel('')
i, k = ax.get_ylim()
ax.set_ylim(i+0.5, k-0.5) #устанавливаем границы вручную
plt.title('Наличие детей и возврат кредита в срок', fontsize = 25)
plt.show()
df_debt_children.sort_values(by = 'процент должников', ascending = False)
```

Наличие детей для женщин является фактором, который увеличивает вероятность задолженности. У мужчин такой зависимости не прослеживается, но независимо от количества детей, у мужчин более высокий процент должников
